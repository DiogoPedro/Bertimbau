```python
import os
import re
from google.colab import drive
import tweepy as tw
import pandas as pd
from cleantext import clean
from random import randrange
```

```python
%cd /content/drive/MyDrive/Token
with open('chaves.txt', 'r') as tfile:
    consumer_key = tfile.readline().strip('\n')
    consumer_secret = tfile.readline().strip('\n')
    access_token = tfile.readline().strip('\n')
    access_token_secret = tfile.readline().strip('\n')
    access_bearer_token = tfile.readline().strip('\n')
```

```python
auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
ListaDeValores = []
count = 0
nome = []
twitter = []
conteudo = []
favorDeLula = []
favorDeBolsonaro = []
nuloAosCandidatos = []

regex = r"((https?):((//)|(\\\\))+[\w\d:#@%/;$()~_?\+-=\\\.&]*)"
```

```python
query_search = "#Lula13"  + " -filter:retweets" 
cursor_tweets = tw.Cursor(api.search_tweets,q=query_search, tweet_mode="extended").items(200)

for tweet in cursor_tweets:

  tweet.full_text = tweet.full_text.replace("@","")
  url = re.findall(regex,tweet.full_text)

  if 7 >= randrange(10):
    tweet.full_text = tweet.full_text.replace("#lula13","lula")
    tweet.full_text = tweet.full_text.replace("#bolsonaro22","bolsonaro")
    tweet.full_text = tweet.full_text.replace("#","")

  if(len(url) == 0) and (len(tweet.full_text) >= 60):
      texto_do_twitter = clean(tweet.full_text, no_emoji=True)
      nome_do_usuario = clean(tweet.user.name, no_emoji=True)

      if texto_do_twitter not in conteudo:
          conteudo.append(texto_do_twitter)
          twitter.append(texto_do_twitter)
          nome.append(nome_do_usuario)
          favorDeLula.append(1)
          favorDeBolsonaro.append(0)
          nuloAosCandidatos.append(0)

      if count <= 10:
        print(tweet)
        count = count + 1

count = 0
```
